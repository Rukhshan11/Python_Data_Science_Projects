# Cleaning car regerteration_year

getting the percentage of vehicles that are not between 1900-2016. 
Determining whether is it safe to remove values outside those boundries.

```python
(~autos["registration_year"].between(1900,2016)).sum() / len(autos.index)
```

It is only 0.04% of data that exists outside of 1900-2016 interval. So, it will be safe to remove data.

```python
autos = autos[autos['registration_year'].between(1900, 2016)]
```

# Exploring price by Vehicle Make

```python
autos['brand'].unique()
```

```python
autos['brand'].value_counts()
```

German brands make top of list with most car sales, 'VW' is almost double of 2nd placed 'BMW'

```python
# Getting the percentage of each brand 
make_perc = autos['brand'].value_counts(normalize = True)
```

```python
make_perc
```

selecting vehicle make with more than 5% of sale.

```python
top_make = make_perc[make_perc > 0.05].index
```

```python
top_make
```

Looping through top brands to find their mean price

```python
mean_price = {}
for make in top_make:
    makes = autos[autos['brand'] == make]
    mean_make = makes['price'].mean()
    mean_price[make] = int(mean_make)
    
mean_price
```

'BMW', 'MERCEDES', 'AUDI' are luxury car makes.
'Volkswagen' is mid-class brand
'Ford' and 'Opel' make budget cars.

```python
mean_mileage = {} 

for make in top_make:
    makes = autos[autos['brand'] == make]
    kmp = makes['odometer_km'].mean()
    mean_mileage[make] = kmp
    
mean_mileage
```

```python
mean_price_series = pd.Series(mean_price) 
```

```python
mean_price_series
```

```python
mean_milage_series = pd.Series(mean_mileage)
```

```python
mean_milage_series
```

```python
df = pd.DataFrame(mean_price_series, columns = ['mean_price'])
```
